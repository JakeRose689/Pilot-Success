```python
import pandas as pd
```

```python
file = 'pilotData.csv'
df = pd.read_csv(file)
df.head()
```

```python
df['Start Month'] = df['Course Start Date'].str[:2].astype(float)
df
```

```python
df2 = df.groupby(['Person Id']).count()
df2
```

```python
df2 = df2[['Start Month']]
df2.rename(columns={'Start Month':'Course Count'}, inplace=True)
df2
```

```python
df3 = pd.merge(df, df2, on=['Person Id'])
df3
```

```python
data = {'Course Short Name': ['1 Intro', '2 Private', '3 Inst', '4 Com Sgl', '5 Com Mul', '6 CFI', '7 CFI Std',
                            '8 CFII', '9 MEI'],
        'Target Hours': [10, 80, 50, 130, 20, 20, 10, 16, 20]
        }

df4 = pd.DataFrame(data)
df4
```

```python
df5 = pd.merge(df3, df4, on=['Course Short Name'])
df5
```

```python
df5['Variance (Hours)'] = df5["Target Hours"] - df5["Course Flight Hours"]
df5
```

```python
data = {'Course Short Name': ['1 Intro', '2 Private', '3 Inst', '4 Com Sgl', '5 Com Mul', '6 CFI', '7 CFI Std',
                            '8 CFII', '9 MEI'],
        'Target Days': [30, 90, 60, 150, 21, 75, 14, 30, 21]
        }

days = pd.DataFrame(data)
days
```

```python
df6 = pd.merge(df5, days, on=['Course Short Name'])
df6
```

```python
df6['Variance (Days)'] = df6["Target Days"] - df6["E2E Course Days"]
df6
```

```python
clean_df = df6[['Course Short Name', 'Start Month', 'Course Flight Hours', 'E2E Course Days', 'Course Count']]
clean_df.to_csv('cleaned_data.csv', index=False)
```
